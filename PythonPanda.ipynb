```python
import pandas as pd

kolommen = ["Name", "Platform", "EU_Sales", "JP_Sales", "NA_Sales"]

dfGamesKolommen = dfGames[kolommen]
dfGamesSorted = dfGamesKolommen[(dfGamesKolommen.Platform == "NES")|(dfGamesKolommen.Platform == "SNES")|(dfGamesKolommen.Platform == "Wii")]
Sales_Europa = dfGamesSorted["EU_Sales"].sum()
Sales_Japan = dfGamesSorted["JP_Sales"].sum()
Sales_NorthAmerica = dfGamesSorted["NA_Sales"].sum()
print("Sales Japan:", Sales_Japan, "Sales Europa:", Sales_Europa, "Sales North Amarica:", Sales_NorthAmerica)
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

dfBijen = pd.read_csv("honingbijen.csv")

map = folium.Map(location=[52.370304,4.918527], zoom_start=10)

kolommen = ["LAT", "LNG"]
dfBijenLocaties = dfBijen[kolommen]
bijenkorvenLijst = dfBijenLocaties.values.tolist()

map.add_child(HeatMap(bijenkorvenLijst))

map
```
